# ¿Qué es Mongo DB?

MongoDB es la base de datos no relacional, que nos permite:

- Aprovechar los datos y la tecnología para maximizar la ventaja competitiva


- Reducir el riesgo de implementaciones de misión crítica


- Acelerar el tiempo de valor drásticamente, menor costo total de propiedad

A diferencia de las bases de datos relacionales, donde se almacenan los registros como filas de una tabla, con MongoDB estos registros son almacenados en documentos BSON (un derivado de los documentos JSON).

Veámos los conceptos básicos presentes en MongoDB: 

- Los `documentos`: cada una de las entradas de nuestra colección. 


- Las `colecciones`: es un conjunto de documentos y podría identificarse con una tabla con la particularidad de que no existe un esquema fijo (es decir, el número de campos y sus nombres pueden variar)


- Las `bases de datos`: conjunto de colecciones



````python
# si pasamos el parámetro de consulta sin niguna especificación, nos devolverá todos los elementos de la BBDD

coleccion.find()
````

### Búsqueda de campos de tipo string 

Ahora ya lo tenemos claro, y queremos sacar todos los alojamientos de la isla de "El Hierro". 

````python
list(coleccion.find({"nom_isla":"El Hierro"}))[:2]
````

### Búsqueda de campos de tipo numérico

Para hacer búsquedas con valores númericos tendremos que usar los operadores de comparación, que en Mongo son un poco diferentes: 

- `$eq`: igual a 


- `$gt`: mayor que


- `$gte`: mayor o igual que


- `$lt`: menor que


- `$lte`: menor o igual que


- `$in`: dentro de


- `$nin`: no dentro de


- `$ne`: distinto de 


1️⃣ `$eq`: igual a 

Busquemos aquellos alojamientos que tengan 3⭐️

````python
list(coleccion.find({'cod_categor': {"$eq": 3}} ))[:2]
````

2️⃣ `$gt`: mayor que 

Busquemos aquellos alojamientos que tengan mas de 3⭐️

````python
mayor = pd.DataFrame(coleccion.find({'cod_categor': {"$gt": 3}}))
mayor.head(2)
````

````python
mayor.nom_categor.unique()
````

3️⃣ `$gte`: mayor o igual que 

Busquemos aquellos alojamientos que tengan 3 o mas ⭐️

````python
mayor_igual = pd.DataFrame(coleccion.find({'cod_categor': {"$gte": 3}}))
mayor_igual.head(2)
````

````python
mayor_igual.nom_categor.unique()
````

4️⃣ `$lt`: menor que 

Busquemos aquellos alojamientos que tengan menos de 3⭐️

````python
menor = pd.DataFrame(coleccion.find({'cod_categor': {"$lt": 3}}))
menor.head(2)
````

````python
menor.nom_categor.unique()
````

5️⃣ `$lte`: menor que 

Busquemos aquellos alojamientos que tengan 3 o menos ⭐️

````python
menor_igual = pd.DataFrame(coleccion.find({'cod_categor': {"$lte": 3}}))
menor_igual.head(2)
````

````python
menor_igual.nom_categor.unique()
````

6️⃣ `in()`: dentro de

Queremos sacar todos los alojamientos que están en Lanzarote y en El Hierro

````python
# creamos una lista con las islas que queremos buscar.

islas = ["Lanzarote", "El Hierro"]
````

````python
list(coleccion.find({"nom_isla": {"$in": islas}}))[:3]
````

7️⃣ `nin()` no dentro de. 

Hace referencia a:

- el valor del campo no está


- el campo no existe


Imaginemos que queremos buscar todas las islas. Podríamos hacer un `find` e indicar el nombre de todas las islas. También podemos usar `nin` y especificar solo el que no queremo

````python
list(coleccion.find({"nom_isla": {"$nin": ["Lanzarote"]}}))[:3]
````

8️⃣ `$ne`: distinto de

Si quisieramos hacer una búsqueda como la anterior también podemos usar. Buscaremos todo lo que sea distinto de Lanzarote

````python
list(coleccion.find({"nom_isla": {"$ne": ["Lanzarote"]}}))[3:6]
````

## Operadores `and` y `or`

Combinemos dos condiciones, aquí podemos usar dos operadores: 

- `$and`: se tienen que cumplir todas las condiciones que le pasamos


- `$or`: que se cumpla una condición u otra

````python
cond1 = {"nom_isla":"El Hierro"}
cond2 = {'cod_categor': {"$eq": 3}}

list(coleccion.find({'$and': [cond1, cond2]}))
````

````python
list(coleccion.find({'$or': [cond1, cond2]}))[5:7]
````

## Proyecciones

Haremos proyecciones cuando no queramos que nos devuelvan todos los campos de nuestro documento.

````python
list(coleccion.find({'$or': [cond1, cond2]}, {"etiqueta":1,"_id":0}))[8:20]
````

## `sort()` 

El sentido del orden los especificaremos pasando un `1` o un `-1`.

- `1`: orden ascendente.


- `-1`: orden descendente.

## `type()` 

Elige los documentos en los que el valor del campo es una instancia del tipo. 

Los tipos que hay en Mongo los tenéis [aquí](https://docs.mongodb.com/manual/reference/operator/query/type/)

````python
# tenemos algún nulo en la campo "nom_tipo"?
list(coleccion.find({"nom_tipo": {"$type": "null"}}, {"_id":0, "etiqueta":1}))
````

````python
# devuelveme todo lo que sea de tipo string del campo "nom_tipo"
list(coleccion.find({"nom_tipo": {"$type": "string"}}, {"_id":0, "etiqueta":1}).limit(4)) 
````

1️⃣ Crear la geometría

````python
def crear_geometria(lon,lat):
    return {"type": "Point", "coordinates":[ lon, lat]}
````

````python
df.head(1)
````

````python
df_geometry = df.copy()
````

**No se podrán hacer los índices más adelante si tenemos NaN, por lo que lo primero vamos a eliminar esos nulos**

````python
df_geometry.dropna(how = "any", inplace = True)
````

````python
df_geometry["geometria"] = df_geometry.apply(lambda datos: crear_geometria(
                                    datos["longitud"],
                                    datos["latitud"]), axis = 1)
````

## Tipos de geoqueries 

- `near`: especifica un punto para el que una consulta geoespacial devuelve los documentos desde el más cercano al más lejano. El operador `$near` puede especificar un punto GeoJSON o un punto de coordenadas heredado.



- `geonear`: muestra los documentos en orden de más cercano a más lejano de un punto especificado. Además nos va a incluir una columna más indicando los metros de distancia que hay entre el punto de referencia y cada uno de los de nuestra colección



### `near()`

Necesitaremos: 

- Unas coordenadas de referencia (en nuestro ejemplo geopoint). Recordamos que tienen que estar en formato de geopunto


- Unos metros máximos o mínimos de distancia (en nuestro ejemplo metros)

````python
# definimos el punto sobre el que queremos hacer nuestra búsqueda

geopoint = {'type': 'Point', 'coordinates':[-15.438072, 28.122201]}
metros = 3000
````

````python
busqueda = coleccion.find({'geometria':{
        '$near': {'$geometry': geopoint, '$maxDistance':metros}
    }})
busqueda
````

````python
resultado = pd.DataFrame(busqueda)
````

````python
resultado.shape
````

````python
resultado.head(2)
````

## `geonear()` 

Es una query de agregación. ¿Qué significa esto? 

La agregación en MongoDB sigue una estructura tipo *pipeline*, es decir,  diferentes etapas, donde cada una toma la salida de la anterior. Los elementos del *pipeline* se incluyen en un array y se ejecutarán por orden. 
Necesitaremos: 

- Unas coordenadas de referencia (en nuestro ejemplo geopoint). Recordamos que tienen que estar en formato de geopunto


- Unos metros máximos o mínimos de distancia (en nuestro ejemplo metros)

````python
query = [{
    "$geoNear": {'near': geopoint, # punto de referencia
                 'distanceField': 'distance', # el campo de salida que contiene la distancia calculada. 
                 'maxDistance': metros, # Opcional. Distancia en metros. La distancia máxima desde el punto central que pueden tener los documentos. 
                 'distanceMultiplier': 6371, # Opcional. El factor para multiplicar todas las distancias devueltas por la consulta. Por ejemplo, utilizamos el multiplicador de distancia para convertir los radianes, devueltos por una consulta esférica, en kilómetros multiplicándolos por el radio de la Tierra.
                 'spherical'  : True}}] # pra que tenga en cuenta que tiene que calcular las distancias en una esfera
geoloc = coleccion.aggregate(query)
````

````python
df_geonear = pd.DataFrame(geoloc)
````

````python
df_geonear.shape
````
